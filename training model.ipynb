```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
def plot_learning_curves(model, X, Y):
    x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2)
    train_error, val_error = [], []
    for m in range(1, len(x_train)):
        model.fit(x_train[:m], y_train[:m])
        y_train_predict = model.predict(x_train[:m])
        y_val_predict = model.predict(x_val)
        train_error.append(mean_squared_error(y_train_predict, y_train[:m]))
        val_error.append(mean_squared_error(y_val_predict, y_val))
    plt.ylim(0,2)
    plt.plot(np.sqrt(train_error), "r-+", linewidth=1, label="train")
    plt.plot(np.sqrt(val_error), "b-", linewidth=1, label="val")
    
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, x_poly, y_poly)
plt.show()
```

```python
from sklearn.base import clone
sgd_reg = SGDRegressor(n_iter=1, warm_start=True, penalty=None, learning_rate="constant", eta0=0.0005) #warm_start=continue training the model from where it pause last time
mini_val_error=float("inf") # initiate error is infinite
best_epoch = None
best_model = None
best_intercept = None
best_coef = None
errors = []
for epoch in range(1000):
    sgd_reg.fit(x_poly,y_poly)
    y_pred = sgd_reg.predict(x_poly)
    val_error = mean_squared_error(y_pred, y_poly)
    errors.append(val_error)
    if val_error < mini_val_error:
        mini_val_error = val_error
        best_intercept = sgd_reg.intercept_
        best_coef = sgd_reg.coef_
        best_epoch = epoch
        best_model = clone(sgd_reg)
plt.ylim(0,8) 
plt.xlim(1,10)
#haha = np.c_[]
fuc = []
for i in range(1,1001):
    fuc.append(i)
plt.plot(fuc,errors)
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
X = iris["data"][:,3:]
Y = (iris["target"]==2).astype(np.int)
Y
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,Y)
X_new = np.linspace(0 ,3, 1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:,1], "b--", label="Iris-Virginica")
plt.plot(X_new, y_proba[:,0], "r--", label="Non Iris-Virginica")
plt.show()
```

```python
log_reg.predict([[1.62]])
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_binary = X_new[y_proba[:,1] >= 0.5][0]
decision_binary
```

```python
plt.figure(figsize=(8,3))
plt.plot(X[Y==0],Y[Y==0], "bs")
plt.plot(X[Y==1],Y[Y==1], "g^")
plt.plot([decision_binary,decision_binary],[0,1], "k:", linewidth=2)
plt.plot(X_new, y_proba[:,1], "g-", linewidth=2, label="iris-vir")
plt.plot(X_new, y_proba[:,0], "b--", linewidth=2, label="non-iris-vir")
plt.text(decision_binary,0.75, "Decision Binary", fontsize=10, color="r", ha="center")
plt.arrow(decision_binary, 0.1, -0.3, 0, head_width=0.05, head_length=0.1, fc="b", ec="b")
plt.arrow(decision_binary, 0.9, 0.3, 0, head_width=0.05, head_length=0.1, fc="g", ec="g")
plt.xlabel("petal width")
plt.ylabel("probability")
plt.legend(loc="center left", fontsize=15)
plt.show()
```
