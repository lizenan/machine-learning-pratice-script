```python
import numpy as np
from matplotlib import pyplot as plt
def add_noise(y):
    """
    Adds random noise to a set of points
    """
    noise_func = np.vectorize(lambda x: x + np.random.uniform(-np.max(y)/0.5, np.max(y)/0.5, 1))#broadcasting
    return noise_func(y)

def generate_linear_points(start, stop, number, m=-1, b=3):
    """
    Generates points along a line with the equation
    y = mx + b (parent function)
    """
    x = np.linspace(start, stop, num=number)
    linear_function = np.vectorize(lambda x: m*x*x + b)
    y = linear_function(x)
    return (x, y)



def get_linear_y_values(X, m=None, b=None):
    return m*X*X + b

X, _Y = generate_linear_points(1, 5, 100, m=-1, b=3)
## Add random noise to the y values
Y = add_noise(_Y)

plt.scatter(X, Y)
plt.show()
```

```python
def get_linear_gradients_mse_m(X, Y, Y_hat, m):
    """
    returns gradients for the two variables m and b assuming mean square error
    for a linear equation y = mx + b
    """
    # dL/dm
    m_gradient = np.sum(2*X**2*(Y_hat-Y))/len(Y) ### Try removing the normalization factor len(Y)
    # dL/db
    b_gradient = np.sum(2*(Y_hat-Y))/len(Y)
    return m_gradient, b_gradient

m = -5
b = 10
# Learning rate
alpha = 0.001
# Number of iterations
N = 10000
for i in range(N):
    Y_hat = get_linear_y_values(X, m, b)
    m_grad, b_grad = get_linear_gradients_mse_m(X, Y, Y_hat, m)
    
    # update m and b for the next iteration
    m = m - alpha * m_grad
    b = b - alpha * b_grad
    #print(m, b)
# The final value of 
plt.plot(X, get_linear_y_values(X, m, b), 'r')
plt.scatter(X, Y)
plt.show()
print ("Estimates after {} iterations: m = {}, b = {}".format(N, m, b))
```
